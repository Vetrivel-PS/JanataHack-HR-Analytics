```python
#inspect df
def inspect_df(data):
    dataframe = data.head()
    datasummary = data.describe()
    datamissing = data.isnull().sum()
    return dataframe,datasummary,datamissing
```

```python
#load train data
train = pd.read_csv(Path+"/train_jqd04QH.csv")
```

```python
test = pd.read_csv(Path+"/test_KaymcHn.csv")
```

```python
train.shape
```

```python
inspect_df(train)
```

```python


train['experience'] = train['experience'].replace('>20', 20)
train['experience'] = train['experience'].replace('<1', 1)

test['experience'] = test['experience'].replace('>20', 20)
test['experience'] = test['experience'].replace('<1', 1)

train['last_new_job'] = train['last_new_job'].replace('>4', 5)
train['last_new_job'] = train['last_new_job'].replace('never', 0)

test['last_new_job'] = test['last_new_job'].replace('>4', 5)
test['last_new_job'] = test['last_new_job'].replace('never', 0)

```

```python

test['last_new_job'] = test['last_new_job'].interpolate(method='polynomial', order= 2)
test['last_new_job'] = test['last_new_job'].fillna(1.0)

# Filling the Missing Values
train['company_type'] = train['company_type'].fillna("MSME")
train['company_size'] = train['company_size'].fillna(50000)
train['gender'] = train['gender'].fillna(method= "pad")
train['major_discipline'] = train['major_discipline'].fillna("Other")
train['education_level'] = train['education_level'].fillna(train['education_level'].mode()[0])
train['enrolled_university'] = train['enrolled_university'].fillna(train['enrolled_university'].mode()[0])

test['company_type'] = test['company_type'].fillna("MSME")
test['company_size'] = test['company_size'].fillna(50000)
test['gender'] = test['gender'].fillna(method= "pad")
test['major_discipline'] = test['major_discipline'].fillna("Other")
test['education_level'] = test['education_level'].fillna(test['education_level'].mode()[0])
test['enrolled_university'] = test['enrolled_university'].fillna(test['enrolled_university'].mode()[0])
```

```python
# Use LabelEncoder

from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()

train['city'] = Encoder.fit_transform(tuple(train['city']))
train['relevent_experience'] = Encoder.fit_transform(tuple(train['relevent_experience']))
train['gender'] = Encoder.fit_transform(tuple(train['gender']))
train['enrolled_university'] = Encoder.fit_transform(tuple(train['enrolled_university']))
train['education_level'] = Encoder.fit_transform(tuple(train['education_level']))
train['major_discipline'] = Encoder.fit_transform(tuple(train['major_discipline']))
train['company_size'] = Encoder.fit_transform(tuple(train['company_size']))
train['company_type'] = Encoder.fit_transform(tuple(train['company_type']))

# Encode for Test Data
test['city'] = Encoder.fit_transform(tuple(test['city']))
test['relevent_experience'] = Encoder.fit_transform(tuple(test['relevent_experience']))
test['gender'] = Encoder.fit_transform(tuple(test['gender']))
test['enrolled_university'] = Encoder.fit_transform(tuple(test['enrolled_university']))
test['education_level'] = Encoder.fit_transform(tuple(test['education_level']))
test['major_discipline'] = Encoder.fit_transform(tuple(test['major_discipline']))
test['company_size'] = Encoder.fit_transform(tuple(test['company_size']))
test['company_type'] = Encoder.fit_transform(tuple(test['company_type']))
```
