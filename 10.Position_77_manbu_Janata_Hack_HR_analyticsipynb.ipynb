```python
for column in ['city', 'gender', 'relevent_experience','enrolled_university', 'education_level', 'major_discipline','experience', 'company_size', 'company_type', 'last_new_job']:
    data[column].fillna(data[column].mode()[0], inplace=True)
for column in ['city_development_index','training_hours']:
    data[column].fillna(data[column].mean(), inplace=True)
    
for column in ['city', 'gender', 'relevent_experience','enrolled_university', 'education_level', 'major_discipline','experience', 'company_size', 'company_type', 'last_new_job']:
    future_data[column].fillna(future_data[column].mode()[0], inplace=True)
for column in ['city_development_index','training_hours']:
    future_data[column].fillna(future_data[column].mean(), inplace=True)
```

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0,2,3,4,5,6,7,8,9,10])], remainder='passthrough')


data_x = np.array(ct.fit_transform(data_x))
print(type(data_x))
future_data = np.array(ct.fit_transform(future_data))
print(future_data)
# Encoding the Dependent Variable


```python
data_x=pd.get_dummies(data_x,columns=['city','gender', 'relevent_experience','enrolled_university', 'education_level', 'major_discipline','experience', 'company_size', 'company_type', 'last_new_job'])
future_data=pd.get_dummies(future_data,columns=['city','gender', 'relevent_experience','enrolled_university', 'education_level', 'major_discipline','experience', 'company_size', 'company_type', 'last_new_job'])
type(data_x)
data_y=np.array(data_y)
print(type(data_y))
print(type(data_x))
```

```python
data_x
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(data_x, data_y)

# Predicting the Test set results
y_pred = classifier.predict_proba(future_data)

print(y_pred)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 186))

# Adding the second hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(data_x, data_y, batch_size = 10, nb_epoch = 10)

# Part 3 - Making the predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict_proba(future_data)
```
